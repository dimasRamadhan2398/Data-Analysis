```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mean = 0
std = 1
X = np.arange(-4,4,0.01)
Y = stats.norm.pdf(X,mean,std)
plt.plot(X,Y)
```

```python
mu,sigma = 0,0.1
normal_set = np.random.normal(mu,sigma,1000)
plt.hist(normal_set,bins=50)
```
